## Steady State

```python
# working directory
work_dir = '../dpsim/build/Logs/'
```

```python
# read log file
logName_orig = 'DP_SynchronGenerator_dq_SteadyState_LeftVector'
print(work_dir + logName_orig + '.csv')
ts_dpsim_orig = read_timeseries_dpsim(work_dir + logName_orig + '.csv')
```

```python
phasors_orig = get_node_voltage_phasors(ts_dpsim_orig)
for node, phasor in phasors_orig.items():
    print(node + ': ' + str(phasor['abs'].values[0]) + '<' + str(phasor['phase'].values[0]))
```

```python
emt_orig = get_node_emt_voltages(ts_dpsim_orig, 60)
```

```python
plot_timeseries(1, emt_orig['node00000'])
plot_timeseries(1, emt_orig['node00001'])
plot_timeseries(1, emt_orig['node00002'])
```
