```python
from dataprocessing.dpsim import *
from dataprocessing.plottools import *
```

## Steady State

```python
# working directory
work_dir = '../../../Logs/'
```

```python
# read log file
logName_ststate = 'DP_SynGen_TrStab_SteadyState_LeftVector'
print(work_dir + logName_ststate + '.csv')
ts_dpsim_ststate = read_timeseries_dpsim(work_dir + logName_ststate + '.csv')
```

```python
phasors_ststate = get_node_voltage_phasors(ts_dpsim_ststate)
for node, phasor in phasors_ststate.items():
    print(node + ': ' + str(phasor['abs'].values[0]) + '<' + str(phasor['phase'].values[0]))
```

```python
emt_ststate = get_node_emt_voltages(ts_dpsim_ststate, 60)
```

```python
plot_timeseries(1, emt_ststate['node00000'])
plot_timeseries(1, emt_ststate['node00001'])
```
