# Dynamic Phasor Basics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 0.1, 1000)
omega = 2*np.pi*50

voltage_slow_re = np.cos(-5*2*np.pi*time)
voltage_slow_im = np.sin(-5*2*np.pi*time)

voltage_fund_re = np.cos(0)
voltage_fund_im = np.sin(0)
voltage_emt = voltage_fund_re*np.cos(omega*time) - voltage_fund_im*np.sin(omega*time)

# Shift dynamic phasors by 50 Hz
voltage_fund = voltage_fund_re * np.cos(omega*time) - \
              voltage_fund_im * np.sin(omega*time)
voltage_slow = voltage_slow_re*np.cos(omega*time) - \
              voltage_slow_im*np.sin(omega*time)
```

```python
# Plot vectors
fig, ax = plt.subplots()
ax.plot(time, voltage_emt, 'm-', label = 'EMT')
ax.plot(time, voltage_fund, 'c--', label = 'DP fund')
ax.plot(time, voltage_slow, 'r--', label = 'DP slow')

# Add legend and labels
ax.legend(loc='lower right')
ax.set_xlabel('time [s]')
ax.set_ylabel('voltage [V]')
```
